## ğŸ¯ What This Script Does(Conceptual Overview)
This script performs a **one-time dataset split** for an image classification project.

The script processes the dataset one disease class at a time and follows the same procedure for each class:
For each class folder:
- Randomly selects images
- Keeps approximately:
  - **70%** in `train`
  - **15%** in `val`
  - **15%** in `test`
- Ensures **no image appears in more than one split**

---
##### **Prevention of Data Leakage**
The script guarantees that:
1. Each image file exists in exactly one split:
- If an image is in train, it does not appear in val or test
- If an image is in val, it is never used for training
- If an image is in test, it is used only for final evaluation

2. so that Images used for evaluation were never seen during training. which enables fair model evaluation.

---
##### **Reproducibility**
When working with image datasets, it is important that the same images are used every time the model is trained or tested.
our script does this, after script is runned once the dataset structure is fixed means it is saved on disk itself.

everytime we run training on this dataset or Re-running training does not reshuffle images
 our model uses same training images everytime we re train . likewise for testing same test images used. there will be no reshuffling of images i.e train images in test, when we re-run multiple times. as our script
Physically places images into train/, val/, and test/ folders.
Each image exists in only one split.
- so that Every training run sees the same training images,
- Every evaluation uses the same validation and test images, same results produced.which makes our dataset reproducable.

---
##### **Why reproducability is important**

If images were split again every time we run notebook:
different images would go into training and testing results would change each run
it would be hard to know if the model actually improved.
By keeping the dataset fixed:
results are consistent,
model comparisons are fair,
mistakes caused by reshuffling are avoided.

```python
RUN_DATASET_SPLIT = False # âš ï¸ change to true ONLY when you want to update my dataset with other plant folders from orginal kaggel dataset.
# if you want to add more plants, make sure you drag and dropped those plant folders from kaggle dataset into your train,val,test folders.
# then set this to true and run the script to update your dataset.

if RUN_DATASET_SPLIT:
    import os
    import random

    BASE_DIR = "   " #  " enter your dataset - folder name that conatins train,val,test folders". here mine is " i.e dataset "

    splits = {
    "train": 0.70,
    "val": 0.15,
    "test": 0.15
    }

# Get class names
    classes = os.listdir(os.path.join(BASE_DIR, "train"))
    classes = [c for c in classes if not c.startswith('.')]

    for cls in classes:
        print(f"Processing class: {cls}")

    # Take image list from train folder
        train_path = os.path.join(BASE_DIR, "train", cls)
        images = [img for img in os.listdir(train_path)
                  if img.lower().endswith(('.jpg', '.png', '.jpeg'))]

        random.shuffle(images)

        total = len(images)
        train_end = int(total * splits["train"])
        val_end = train_end + int(total * splits["val"])

        keep = {
            "train": set(images[:train_end]),
            "val": set(images[train_end:val_end]),
            "test": set(images[val_end:])
        }

    # Clean each split
        for split in ["train", "val", "test"]:
            split_path = os.path.join(BASE_DIR, split, cls)
            for img in os.listdir(split_path):
                if img not in keep[split]:
                    os.remove(os.path.join(split_path, img))

    print("âœ… Dataset split completed successfully!")

else:
    print("â„¹ï¸ Dataset split skipped (script runned already, this script was just a reference to see how script's split done. To update dataset rerun the script by setting True).")
```
